## Investigation Overview

In this investigation, I wanted to look at the features of students that could be used to predict grade percentage. The main focus was on gender, parents level of education, and test preparation course.

## Dataset Overview

This data set includes scores from three exams and a variety of personal, social, and economic factors that have interaction effects upon them.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
students = pd.read_csv("Student_newfeatures.csv",index_col=0)
```

## Distribution of Percetanges 

The grade percentages in Students dataset range from 9% to 100%, the distribution of percentages takes a unimodal shape, skewed a little to the left



```python
plt.figure(figsize=[8, 5])
plt.hist(data = students, x = 'Percentage' )
plt.title('Distribution of Percentages')


plt.xlabel('Percentage (%)')
plt.show()
```

## Scores vs Gender

Violin plots that show the differences that gender make in writing, reading, math scores

```python
sb.set_context('notebook', font_scale= 1.3)
fig, ax = plt.subplots(1, 3, figsize = (20, 6))
ax1 = sb.violinplot(x = students['gender'], y = students['writing score'], ax= ax[0])
ax1.set_title("Gender effect on writing score")
ax1 = sb.violinplot(x = students['gender'], y = students['reading score'], ax= ax[1])
ax1.set_title("Gender effect on reading score")
ax1 = sb.violinplot(x = students['gender'], y = students['math score'], ax= ax[2])
ax1.set_title("Gender effect on math score")
plt.show()
```

## Percentange and score vs parental level of education

Plotting a horizental barplot that shows the difference in percentage and scores across different parental levels of education

```python

students.groupby('parental level of education').agg('mean').sort_values(by = 'Percentage').plot(kind='barh',figsize=(15,10))
plt.title("Parental level of education effect on students' scores and percentage")
plt.xlabel("Score/Percetange (%)")
plt.legend(bbox_to_anchor=(1.03, 1), loc = 2);
```

## Percentange and test preparation course

Plotting a KDEplot showing how the preparation course impact the grade Percentage

```python

sb.kdeplot(data = students, x = "Percentage", hue = "test preparation course", cumulative = True, common_norm = False)
plt.title("Distribution of Percentage based ob preparation course")
plt.show()
```
